#Fake News Headline Generator

```python
import random

subjects = [
    "Shahrukh Khan",
    "Virat Kohli",
    "Nirmala Sitharaman",
    "A Mumbai Cat",
    "Narendra Modi",
    "Crime Master Gogo",
    "Auto Rickshaw Wala",
]

actions = [
    "launches",
    "cancels",
    "eats",
    "declares war",
    "orders",
    "celebrates",
    "dances with",
    "sings with"
]

places_or_things = [
    "at Red Fort",
    "inside Mumbai local",
    "in Shaniwar Wada",
    "in hospital",
    "in stadium",
    "in Kashmir valley",
    "on Goda Ghat"
]

# Start loop
while True:
    subject = random.choice(subjects)
    action = random.choice(actions)
    place_or_thing = random.choice(places_or_things)

    headline = f"BREAKING NEWS: {subject} {action} {place_or_thing}"
    print("\n" + headline)

    user_input = input("\nDo you want another headline? (yes/no): ").strip().lower()

    if user_input == "no":
        break

print("\nThanks for using fake news headline generator 😄")
```
